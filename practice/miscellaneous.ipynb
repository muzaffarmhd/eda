```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({
    'timestamp':[0,0.2,1.2,2,2.5,3],
    'temperature':[27.3,23.4,25.6,28.4,22.3,25.1]
})
df
```

```python
df.plot(x='timestamp',y='temperature', kind='line', marker='o')
mean_value = df['temperature'].mean()
plt.axhline(y=mean_value, linestyle='--', label=f'Mean = {mean_value:.2f}', color='red')
print(f'Mean: {mean_value}')
plt.show()
```

```python
df.plot(x='timestamp',y='temperature', kind='line', marker='o')
median_value = df['temperature'].median()
plt.axhline(y=median_value, linestyle='--', label=f'Mean = {median_value:.2f}', color='red')
plt.show()
```

```python
# let's see the effect for outliers or extreme values
df['temperature'] = [77.3,23.4,25.6,28.4,22.3,70]
df.plot(x='timestamp',y='temperature', kind='line', marker='o')
mean_value = df['temperature'].mean()
plt.axhline(y=mean_value, linestyle='--', label=f'Mean = {mean_value:.2f}', color='red')
plt.show()
```

```python
# let's see the effect for outliers or extreme values, we can see that the mean is very sensitive to extreme values, but median isn't, we can use a trimmed
# mean as well for extreme values
df['temperature'] = [77.3,23.4,25.6,28.4,22.3,70]
df.plot(x='timestamp',y='temperature', kind='line', marker='o')
median_value = df['temperature'].median()
plt.axhline(y=median_value, linestyle='--', label=f'Median = {median_value:.2f}', color='red')
print(f"meadian: {median_value}")
plt.show()
```
